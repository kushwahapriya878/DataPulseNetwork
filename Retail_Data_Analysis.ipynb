The dataset contains 302,010 entries and 30 columns with various types of information related to retail transactions. Here's a brief overview of some key columns:
*  Transaction_ID: Unique identifier for each transaction.
*  Customer_ID: Unique identifier for each customer.
Name, Email, Phone, Address, City, State, Zipcode, Country: Customer contact details.
*  Age, Gender, Income, Customer_Segment: Customer demographic information.
*  Date, Year, Month, Time: Transaction date and time details.
*  Total_Purchases, Amount, Total_Amount: Purchase and transaction amount details.
*  Product_Category, Product_Brand, Product_Type: Information about the products purchased.
*  Feedback, Shipping_Method, Payment_Method, Order_Status, Ratings, products: Transaction and product feedback details.

```python
df.dropna(subset='Transaction_ID', inplace=True)
df['Customer_ID']=df['Customer_ID'].fillna('Unknown')
df['Name']=df['Name'].fillna('Unknown')
df['Email']=df['Email'].fillna('Unknown')
df['Address']=df['Address'].fillna('Unknown')
```

```python
df.isna().sum()
```

```python
# converting data type for the following columns
df['Transaction_ID']= df['Transaction_ID'].astype(int)
df['Customer_ID']= df['Customer_ID'].astype(object)
df['Phone']= df['Phone'].astype(object)
df['Zipcode']= df['Zipcode'].astype(object)
df['Date']= pd.to_datetime(df['Date'])
df['Year']= pd.to_datetime(df['Year'])
df['Time']= pd.to_datetime(df['Time'])
df['Total_Purchases']= df['Total_Purchases'].astype(int)
df['Ratings']= df['Ratings'].astype(int)
```
